```python

#Normalizar  La columna valoración

# Extraer solo la puntuación numérica de la columna 'valoración'
df_listings_ciudades2024_normalizado['valoración'] = df_listings_ciudades2024_normalizado['valoración'].str.extract(r'(\d+,\d+|\d+\.\d+)')
# Nota: La expresión regular r'(\d+,\d+|\d+\.\d+)' busca coincidencias de números que contengan un punto o una coma decimal.


# Renombrar la columna 'valoración' a 'rating'
df_listings_ciudades2024_normalizado.rename(columns={'valoración': 'rating'}, inplace=True)
```

```python
#Normalizar La columna características

# Dividir la columna 'caracteristicas' en nuevas columnas usando ' · ' como delimitador
new_columns_1 = df_listings_ciudades2024_normalizado['caracteristicas'].str.split(' · ', expand=True)

# # Verificar el número de columnas generadas y sus primeras filas para entender los datos
# print(new_columns_1.head())

import pandas as pd
import numpy as np

# Inicializar las nuevas columnas con valores NaN
df_listings_ciudades2024_normalizado['viajeros'] = np.nan
df_listings_ciudades2024_normalizado['dormitorios'] = np.nan
df_listings_ciudades2024_normalizado['camas'] = np.nan
df_listings_ciudades2024_normalizado['baños'] = np.nan

# Iterar sobre cada fila del DataFrame original
# Iterar sobre cada fila del DataFrame original
for index, row in df_listings_ciudades2024_normalizado.iterrows():
    partes = row['caracteristicas'].replace("[", "").replace("]", "").split(' · ')
    for parte in partes:
        if 'viajero' in parte:
            numero_viajeros = parte.split()[0].replace("'", "")  # Elimina comillas simples
            # Asignar valor después de manejar potenciales valores no numéricos
            df_listings_ciudades2024_normalizado.at[index, 'viajeros'] = int(float(numero_viajeros)) if numero_viajeros.isdigit() else np.nan
        elif 'dormitorio' in parte:
            numero_dormitorios = parte.split()[0].replace("'", "")  # Elimina comillas simples
            # Asignar valor después de manejar potenciales valores no numéricos
            df_listings_ciudades2024_normalizado.at[index, 'dormitorios'] = int(float(numero_dormitorios)) if numero_dormitorios.isdigit() else np.nan
        elif 'cama' in parte:
            numero_camas = parte.split()[0].replace("'", "")  # Elimina comillas simples
            # Asignar valor después de manejar potenciales valores no numéricos
            df_listings_ciudades2024_normalizado.at[index, 'camas'] = int(float(numero_camas)) if numero_camas.isdigit() else np.nan
        elif 'baño' in parte or 'baños' in parte:
            numero_baños = parte.split()[0].replace("'", "").replace(",", ".")  # Elimina comillas simples y ajusta la coma a punto
            # Asignar valor después de manejar potenciales valores no numéricos
            if numero_baños.replace('.', '', 1).isdigit():
                df_listings_ciudades2024_normalizado.at[index, 'baños'] = float(numero_baños)
            else:
                df_listings_ciudades2024_normalizado.at[index, 'baños'] = np.nan

# Verificar el resultado
print(df_listings_ciudades2024_normalizado.head())
```

Este código añade las columnas viajeros, dormitorios, camas, y baños al DataFrame, y cada celda en estas nuevas columnas se inicializa con NaN. Ahora, el DataFrame está listo para que, a medida que iteres sobre cada fila y analices los datos de la columna 'caracteristicas', puedas llenar estas nuevas columnas con información específica extraída.

Cosas a tener en cuenta al extraer los datos de la columna caracterísitcas[]

**Uso de replace para limpiar corchetes:** Para limpiar los corchetes es esencial considerar tanto el texto como los elementos que podrían estar separados por comas dentro de los mismos. Es importante asegurarse de que la cadena 'caracteristicas' esté siendo limpiada correctamente antes de proceder con el split. Esto garantizará que la manipulación de la cadena capture adecuadamente todos los elementos relevantes sin omitir detalles importantes.

**Uso de split()[0] para extraer números:** Esta técnica es efectiva cuando el número se encuentra al inicio de la cadena. Sin embargo, es crucial verificar que no existan espacios o caracteres adicionales antes del número que puedan afectar la extracción correcta. Esta precaución asegura la precisión en la obtención de los datos numéricos relevantes.

**Consideración de plural y singular:** Es importante manejar adecuadamente tanto las formas singulares como las plurales de las palabras clave, como "baño" y "baños". Esto asegura una correcta normalización de los datos, independientemente de la variabilidad en la descripción de las características.

Conversión de tipos de datos: Al asignar los números extraídos a las columnas del DataFrame, se debe tener en cuenta que estos valores son inicialmente cadenas. Convertirlos a un tipo numérico facilitará el procesamiento y análisis subsiguiente, permitiendo un manejo más efectivo de los datos.

```python
#Utilizaremos una función que verifique cada condición de manera explícita y aplicaremos esta función a cada fila del DataFrame mediante el método .apply() para evitar cualquier malentendido en la lógica de clasificación.
def determinar_tipo_alojamiento(name):
    name = name.lower()  # Convertir a minúsculas

    # Verificar cada tipo de alojamiento en orden de especificidad
    if 'habitación compartida' in name:
        return 'Shared room'
    elif 'habitación de hotel' in name or 'hotel' in name or 'hostal' in name:
        return 'Hotel room'
    elif 'habitación en' in name or 'habitación privada' in name or 'bed' in name:
        return 'Private room'
    elif ('apartamento' in name or 'villa' in name or 'vivienda' in name or 'loft' in name or 'casa' in name or 'apto' in name
          or 'barco' in name  or 'adosado' in name or 'alojamiento' in name or 'suite' in name or 'chalet' in name or 'choza' in name
          or 'granja' in name or 'bungalow' in name or 'caravana' in name or 'isla' in name  or 'yurta' in name  or 'cabaña' in name or 'tienda' in name):
        return 'Entire home/apt'
    else:
        return 'Other'  # Para cualquier otro caso que no encaje en las categorías anteriores

# Aplicar la función a cada fila en la columna 'name' para determinar el 'room_type'
df_listings_ciudades2024_normalizado['room_type'] = df_listings_ciudades2024_normalizado['name'].apply(determinar_tipo_alojamiento)

# Verificar el resultado
print(df_listings_ciudades2024_normalizado[['name', 'room_type']].head())
```

```python
#@title Analisis Estadístico de variables

# Ahora puedes recalcular tus estadísticas
variables = ['price', 'beds','bedrooms','bathrooms',]
for var in variables:
    if df_listings_ciudades2024_normalizado[var].dtype == 'float64' or df_listings_ciudades2024_normalizado[var].dtype == 'int64':
        mean = round(df_listings_ciudades2024_normalizado[var].mean(), 2)
        median = round(df_listings_ciudades2024_normalizado[var].median(), 2)
        std_dev = round(df_listings_ciudades2024_normalizado[var].std(), 2)
        mad = round(df_listings_ciudades2024_normalizado[var].mad(), 2)
        percentiles = df_listings_ciudades2024_normalizado[var].quantile([0.1, 0.25, 0.5, 0.75, 0.9, 0.95, 0.99]).round(2)
        rango = df_listings_ciudades2024_normalizado[var].max() - df_listings_ciudades2024_normalizado[var].min()
        IQR = df_listings_ciudades2024_normalizado[var].quantile(0.75) - df_listings_ciudades2024_normalizado[var].quantile(0.25)
        lower_bound = df_listings_ciudades2024_normalizado[var].quantile(0.25) - 1.5 * IQR
        upper_bound = df_listings_ciudades2024_normalizado[var].quantile(0.75) + 1.5 * IQR
        max_value = df_listings_ciudades2024_normalizado[var].max()
        min_value = df_listings_ciudades2024_normalizado[var].min()

        print(f'Variable: {var}')
        print(f'Media: {mean}')
        print(f'Mediana: {median}')
        print(f'Desviación estándar: {std_dev}')
        print(f'Desviación Absoluta Media (MAD): {mad}')
        print(f'Percentiles: {percentiles}')
        print(f"Rango: {rango}")
        print(f"IQR: {IQR}")
        print(f"Límites: [{lower_bound}, {upper_bound}]")
        print(f"Valor máximo: {max_value}")
        print(f"Valor mínimo: {min_value}\n")
    else:
        print(f"Variable: {var} no es numérica y se ha omitido de las estadísticas.\n")
```

Las propiedades muestran una gran variabilidad en el precio, con una tendencia hacia precios más bajos pero con algunos valores extremadamente altos. La cantidad de camas, dormitorios y baños tiende a concentrarse en rangos más estrechos, con la mayoría de las propiedades ofreciendo 1 a 3 dormitorios y baños, y 1 a 4 camas. Los límites calculados para identificar valores atípicos sugieren la presencia de propiedades únicas tanto en términos de precio como de características, lo cual es crucial para tomar decisiones informadas en el mercado inmobiliario o de alquileres.

#### Outliers

```python
#@title Price
# Calcular IQR para 'price'
Q1 = df_listings_ciudades2024_normalizado['price'].quantile(0.25)
Q3 = df_listings_ciudades2024_normalizado['price'].quantile(0.75)
IQR = Q3 - Q1

# Definir límites para identificar outliers
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python

# Identificar outliers
outliers_price = df_listings_ciudades2024_normalizado[(df_listings_ciudades2024_normalizado['price'] < lower_limit) | (df_listings_ciudades2024_normalizado['price'] > upper_limit)]

#Nota =  Este código  mostrará los IDs y precios de los alojamientos que son considerados outliers según el criterio del IQR.


#Visualización

# Boxplot de 'price'
px.box(df_listings_ciudades2024_normalizado, x='city', y='price', title='Boxplot de Precios por Ciudad con outliers')


```
