```python

#Normalizar  La columna valoraci√≥n

# Extraer solo la puntuaci√≥n num√©rica de la columna 'valoraci√≥n'
df_listings_ciudades2024_normalizado['valoraci√≥n'] = df_listings_ciudades2024_normalizado['valoraci√≥n'].str.extract(r'(\d+,\d+|\d+\.\d+)')
# Nota: La expresi√≥n regular r'(\d+,\d+|\d+\.\d+)' busca coincidencias de n√∫meros que contengan un punto o una coma decimal.


# Renombrar la columna 'valoraci√≥n' a 'rating'
df_listings_ciudades2024_normalizado.rename(columns={'valoraci√≥n': 'rating'}, inplace=True)
```

```python
#Normalizar La columna caracter√≠sticas

# Dividir la columna 'caracteristicas' en nuevas columnas usando ' ¬∑ ' como delimitador
new_columns_1 = df_listings_ciudades2024_normalizado['caracteristicas'].str.split(' ¬∑ ', expand=True)

# # Verificar el n√∫mero de columnas generadas y sus primeras filas para entender los datos
# print(new_columns_1.head())

import pandas as pd
import numpy as np

# Inicializar las nuevas columnas con valores NaN
df_listings_ciudades2024_normalizado['viajeros'] = np.nan
df_listings_ciudades2024_normalizado['dormitorios'] = np.nan
df_listings_ciudades2024_normalizado['camas'] = np.nan
df_listings_ciudades2024_normalizado['ba√±os'] = np.nan

# Iterar sobre cada fila del DataFrame original
# Iterar sobre cada fila del DataFrame original
for index, row in df_listings_ciudades2024_normalizado.iterrows():
    partes = row['caracteristicas'].replace("[", "").replace("]", "").split(' ¬∑ ')
    for parte in partes:
        if 'viajero' in parte:
            numero_viajeros = parte.split()[0].replace("'", "")  # Elimina comillas simples
            # Asignar valor despu√©s de manejar potenciales valores no num√©ricos
            df_listings_ciudades2024_normalizado.at[index, 'viajeros'] = int(float(numero_viajeros)) if numero_viajeros.isdigit() else np.nan
        elif 'dormitorio' in parte:
            numero_dormitorios = parte.split()[0].replace("'", "")  # Elimina comillas simples
            # Asignar valor despu√©s de manejar potenciales valores no num√©ricos
            df_listings_ciudades2024_normalizado.at[index, 'dormitorios'] = int(float(numero_dormitorios)) if numero_dormitorios.isdigit() else np.nan
        elif 'cama' in parte:
            numero_camas = parte.split()[0].replace("'", "")  # Elimina comillas simples
            # Asignar valor despu√©s de manejar potenciales valores no num√©ricos
            df_listings_ciudades2024_normalizado.at[index, 'camas'] = int(float(numero_camas)) if numero_camas.isdigit() else np.nan
        elif 'ba√±o' in parte or 'ba√±os' in parte:
            numero_ba√±os = parte.split()[0].replace("'", "").replace(",", ".")  # Elimina comillas simples y ajusta la coma a punto
            # Asignar valor despu√©s de manejar potenciales valores no num√©ricos
            if numero_ba√±os.replace('.', '', 1).isdigit():
                df_listings_ciudades2024_normalizado.at[index, 'ba√±os'] = float(numero_ba√±os)
            else:
                df_listings_ciudades2024_normalizado.at[index, 'ba√±os'] = np.nan

# Verificar el resultado
print(df_listings_ciudades2024_normalizado.head())
```

Este c√≥digo a√±ade las columnas viajeros, dormitorios, camas, y ba√±os al DataFrame, y cada celda en estas nuevas columnas se inicializa con NaN. Ahora, el DataFrame est√° listo para que, a medida que iteres sobre cada fila y analices los datos de la columna 'caracteristicas', puedas llenar estas nuevas columnas con informaci√≥n espec√≠fica extra√≠da.

Cosas a tener en cuenta al extraer los datos de la columna caracter√≠sitcas[]

**Uso de replace para limpiar corchetes:** Para limpiar los corchetes es esencial considerar tanto el texto como los elementos que podr√≠an estar separados por comas dentro de los mismos. Es importante asegurarse de que la cadena 'caracteristicas' est√© siendo limpiada correctamente antes de proceder con el split. Esto garantizar√° que la manipulaci√≥n de la cadena capture adecuadamente todos los elementos relevantes sin omitir detalles importantes.

**Uso de split()[0] para extraer n√∫meros:** Esta t√©cnica es efectiva cuando el n√∫mero se encuentra al inicio de la cadena. Sin embargo, es crucial verificar que no existan espacios o caracteres adicionales antes del n√∫mero que puedan afectar la extracci√≥n correcta. Esta precauci√≥n asegura la precisi√≥n en la obtenci√≥n de los datos num√©ricos relevantes.

**Consideraci√≥n de plural y singular:** Es importante manejar adecuadamente tanto las formas singulares como las plurales de las palabras clave, como "ba√±o" y "ba√±os". Esto asegura una correcta normalizaci√≥n de los datos, independientemente de la variabilidad en la descripci√≥n de las caracter√≠sticas.

Conversi√≥n de tipos de datos: Al asignar los n√∫meros extra√≠dos a las columnas del DataFrame, se debe tener en cuenta que estos valores son inicialmente cadenas. Convertirlos a un tipo num√©rico facilitar√° el procesamiento y an√°lisis subsiguiente, permitiendo un manejo m√°s efectivo de los datos.

```python
#Utilizaremos una funci√≥n que verifique cada condici√≥n de manera expl√≠cita y aplicaremos esta funci√≥n a cada fila del DataFrame mediante el m√©todo .apply() para evitar cualquier malentendido en la l√≥gica de clasificaci√≥n.
def determinar_tipo_alojamiento(name):
    name = name.lower()  # Convertir a min√∫sculas

    # Verificar cada tipo de alojamiento en orden de especificidad
    if 'habitaci√≥n compartida' in name:
        return 'Shared room'
    elif 'habitaci√≥n de hotel' in name or 'hotel' in name or 'hostal' in name:
        return 'Hotel room'
    elif 'habitaci√≥n en' in name or 'habitaci√≥n privada' in name or 'bed' in name:
        return 'Private room'
    elif ('apartamento' in name or 'villa' in name or 'vivienda' in name or 'loft' in name or 'casa' in name or 'apto' in name
          or 'barco' in name  or 'adosado' in name or 'alojamiento' in name or 'suite' in name or 'chalet' in name or 'choza' in name
          or 'granja' in name or 'bungalow' in name or 'caravana' in name or 'isla' in name  or 'yurta' in name  or 'caba√±a' in name or 'tienda' in name):
        return 'Entire home/apt'
    else:
        return 'Other'  # Para cualquier otro caso que no encaje en las categor√≠as anteriores

# Aplicar la funci√≥n a cada fila en la columna 'name' para determinar el 'room_type'
df_listings_ciudades2024_normalizado['room_type'] = df_listings_ciudades2024_normalizado['name'].apply(determinar_tipo_alojamiento)

# Verificar el resultado
print(df_listings_ciudades2024_normalizado[['name', 'room_type']].head())
```

```python
#@title Analisis Estad√≠stico de variables

# Ahora puedes recalcular tus estad√≠sticas
variables = ['price', 'beds','bedrooms','bathrooms',]
for var in variables:
    if df_listings_ciudades2024_normalizado[var].dtype == 'float64' or df_listings_ciudades2024_normalizado[var].dtype == 'int64':
        mean = round(df_listings_ciudades2024_normalizado[var].mean(), 2)
        median = round(df_listings_ciudades2024_normalizado[var].median(), 2)
        std_dev = round(df_listings_ciudades2024_normalizado[var].std(), 2)
        mad = round(df_listings_ciudades2024_normalizado[var].mad(), 2)
        percentiles = df_listings_ciudades2024_normalizado[var].quantile([0.1, 0.25, 0.5, 0.75, 0.9, 0.95, 0.99]).round(2)
        rango = df_listings_ciudades2024_normalizado[var].max() - df_listings_ciudades2024_normalizado[var].min()
        IQR = df_listings_ciudades2024_normalizado[var].quantile(0.75) - df_listings_ciudades2024_normalizado[var].quantile(0.25)
        lower_bound = df_listings_ciudades2024_normalizado[var].quantile(0.25) - 1.5 * IQR
        upper_bound = df_listings_ciudades2024_normalizado[var].quantile(0.75) + 1.5 * IQR
        max_value = df_listings_ciudades2024_normalizado[var].max()
        min_value = df_listings_ciudades2024_normalizado[var].min()

        print(f'Variable: {var}')
        print(f'Media: {mean}')
        print(f'Mediana: {median}')
        print(f'Desviaci√≥n est√°ndar: {std_dev}')
        print(f'Desviaci√≥n Absoluta Media (MAD): {mad}')
        print(f'Percentiles: {percentiles}')
        print(f"Rango: {rango}")
        print(f"IQR: {IQR}")
        print(f"L√≠mites: [{lower_bound}, {upper_bound}]")
        print(f"Valor m√°ximo: {max_value}")
        print(f"Valor m√≠nimo: {min_value}\n")
    else:
        print(f"Variable: {var} no es num√©rica y se ha omitido de las estad√≠sticas.\n")
```

Las propiedades muestran una gran variabilidad en el precio, con una tendencia hacia precios m√°s bajos pero con algunos valores extremadamente altos. La cantidad de camas, dormitorios y ba√±os tiende a concentrarse en rangos m√°s estrechos, con la mayor√≠a de las propiedades ofreciendo 1 a 3 dormitorios y ba√±os, y 1 a 4 camas. Los l√≠mites calculados para identificar valores at√≠picos sugieren la presencia de propiedades √∫nicas tanto en t√©rminos de precio como de caracter√≠sticas, lo cual es crucial para tomar decisiones informadas en el mercado inmobiliario o de alquileres.

```python
#@title Price
# Calcular IQR para 'price'
Q1 = df_listings_ciudades2024_normalizado['price'].quantile(0.25)
Q3 = df_listings_ciudades2024_normalizado['price'].quantile(0.75)
IQR = Q3 - Q1

# Definir l√≠mites para identificar outliers
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python

# Identificar outliers
outliers_price = df_listings_ciudades2024_normalizado[(df_listings_ciudades2024_normalizado['price'] < lower_limit) | (df_listings_ciudades2024_normalizado['price'] > upper_limit)]

#Nota =  Este c√≥digo  mostrar√° los IDs y precios de los alojamientos que son considerados outliers seg√∫n el criterio del IQR.


#Visualizaci√≥n

# Boxplot de 'price'
px.box(df_listings_ciudades2024_normalizado, x='city', y='price', title='Boxplot de Precios por Ciudad con outliers')


```
