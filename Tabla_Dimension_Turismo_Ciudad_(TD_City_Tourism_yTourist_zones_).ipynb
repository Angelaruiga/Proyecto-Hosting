```python

#@title Crear la tabla dinámica para viajeros_y_pernoctaciones_provincias

# Primero, asegúrarnos de que los datos son numéricos.
viajeros_y_pernoctaciones_provincias['viajeros_y_pernoctaciones_provincias'] = pd.to_numeric(viajeros_y_pernoctaciones_provincias[' viajeros_y_pernoctaciones_provincias'], errors='coerce')

# Dividir el DataFrame en cuatro según el tipo de 'Viajeros y pernoctaciones' y 'Residencia'
df_viajeros_espana = viajeros_y_pernoctaciones_provincias[(viajeros_y_pernoctaciones_provincias['Viajeros y pernoctaciones'] == 'Viajero') & (viajeros_y_pernoctaciones_provincias['Residencia'] == 'Residentes en España')]
df_viajeros_extranjero = viajeros_y_pernoctaciones_provincias[(viajeros_y_pernoctaciones_provincias['Viajeros y pernoctaciones'] == 'Viajero') & (viajeros_y_pernoctaciones_provincias['Residencia'] == 'Residentes en el Extranjero')]
df_pernoctaciones_espana = viajeros_y_pernoctaciones_provincias[(viajeros_y_pernoctaciones_provincias['Viajeros y pernoctaciones'] == 'Pernoctaciones') & (viajeros_y_pernoctaciones_provincias['Residencia'] == 'Residentes en España')]
df_pernoctaciones_extranjero = viajeros_y_pernoctaciones_provincias[(viajeros_y_pernoctaciones_provincias['Viajeros y pernoctaciones'] == 'Pernoctaciones') & (viajeros_y_pernoctaciones_provincias['Residencia'] == 'Residentes en el Extranjero')]

# Creamos tablas dinámicas para sumar los valores para cada tipo
pivot_viajeros_espana = df_viajeros_espana.pivot_table(index=['Provincias', 'Region', 'Año'],
                                                        values='viajeros_y_pernoctaciones_provincias',
                                                        aggfunc='sum').rename(columns={'viajeros_y_pernoctaciones_provincias': 'viajeros_nacionales'})

pivot_viajeros_extranjero = df_viajeros_extranjero.pivot_table(index=['Provincias', 'Region', 'Año'],
                                                                values='viajeros_y_pernoctaciones_provincias',
                                                                aggfunc='sum').rename(columns={'viajeros_y_pernoctaciones_provincias': 'viajeros_extranjeros'})

pivot_pernoctaciones_espana = df_pernoctaciones_espana.pivot_table(index=['Provincias', 'Region', 'Año'],
                                                                    values='viajeros_y_pernoctaciones_provincias',
                                                                    aggfunc='sum').rename(columns={'viajeros_y_pernoctaciones_provincias': 'pernoctaciones_nacionales'})

pivot_pernoctaciones_extranjero = df_pernoctaciones_extranjero.pivot_table(index=['Provincias', 'Region', 'Año'],
                                                                            values='viajeros_y_pernoctaciones_provincias',
                                                                            aggfunc='sum').rename(columns={'viajeros_y_pernoctaciones_provincias': 'pernoctaciones_extranjeros'})

# Unimos las cuatro tablas dinámicas en una sola DataFrame final#usamos pd.concat al combinar varias tablas
df_viajeros_pernoctaciones = pd.concat([
    pivot_viajeros_espana, pivot_viajeros_extranjero,
    pivot_pernoctaciones_espana, pivot_pernoctaciones_extranjero
], axis=1).reset_index()


# Redondear la columna 'domestic_travelers' a dos decimales
df_viajeros_pernoctaciones['viajeros_nacionales'] = df_viajeros_pernoctaciones['viajeros_nacionales'].round(2)

df_viajeros_pernoctaciones['viajeros_extranjeros'] = df_viajeros_pernoctaciones['viajeros_extranjeros'].round(2)

df_viajeros_pernoctaciones['pernoctaciones_nacionales'] = df_viajeros_pernoctaciones['pernoctaciones_nacionales'].round(2)
df_viajeros_pernoctaciones['pernoctaciones_extranjeros'] = df_viajeros_pernoctaciones['pernoctaciones_extranjeros'].round(2)

df_viajeros_pernoctaciones.head()  # Muestra las primeras filas del DataFrame final.
```
