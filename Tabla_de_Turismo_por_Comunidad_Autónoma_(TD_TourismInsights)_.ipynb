```python
#@title Crear la tabla dinámica para Total Turistas

# Convertir la columna "Total" a numérico, asumiendo que los datos ya están en el DataFrame `total_turistas`
total_turistas['Total'] = pd.to_numeric(total_turistas['Total'].str.replace('.', '').str.replace(',', '.'), errors='coerce')

# Primero, filtramos para cada tipo y creamos las columnas correspondientes.

df_turistas = total_turistas[total_turistas['Tipo de dato'] == 'Dato base']

df_tasa_variacion = total_turistas[total_turistas['Tipo de dato'] == 'Tasa de variación anual']


# Ahora, creamos la tabla dinámica para cada tipo.

pivot_turistas = df_turistas.pivot_table(index=['Año', 'Region'],
                                           values='Total',
                                           aggfunc='sum').rename(columns={'Total': 'num_turistas'})

pivot_tasa_variacion = df_tasa_variacion.pivot_table(index=['Año', 'Region'],
                                           values='Total',
                                           aggfunc='sum').rename(columns={'Total': 'tasa_variacion'})


# Unimos las cuatro tablas dinámicas en una sola DataFrame final#usamos pd.concat al combinar varias tablas
df_total_turistas   = pd.concat([
    pivot_turistas, pivot_tasa_variacion,
], axis=1).reset_index()


# Redondear la columna 'df_total_turistas' a dos decimales
df_total_turistas['num_turistas'] = df_total_turistas['num_turistas'].round(2)

df_total_turistas['tasa_variacion'] = df_total_turistas['tasa_variacion'].round(2)


df_total_turistas
```

```python
#@title Crear la tabla dinámica para motivo_viaje

# Convertir la columna "Total" a numérico, manejando correctamente los separadores
# motivo_viaje['Total'] = pd.to_numeric(motivo_viaje['Total'].str.replace('.', '').str.replace(',', '.'), errors='coerce')

# No necesitamos filtrar por "Tipo de dato" ya que trabajaremos directamente con los motivos

# Crear tablas dinámicas para cada motivo del viaje
pivot_negocios = motivo_viaje[motivo_viaje['Motivo del viaje'] == 'Negocios y otros motivos profesionales'].pivot_table(index=['Año', 'Region'],
                                                                                                                         values='Total',
                                                                                                                         aggfunc='sum').rename(columns={'Total': 'total_negocios'})

pivot_ocio = motivo_viaje[motivo_viaje['Motivo del viaje'] == 'Ocio, recreo y vacaciones'].pivot_table(index=['Año', 'Region'],
                                                                                                       values='Total',
                                                                                                       aggfunc='sum').rename(columns={'Total': 'total_ocio'})

pivot_otros = motivo_viaje[motivo_viaje['Motivo del viaje'] == 'Otros motivos'].pivot_table(index=['Año', 'Region'],
                                                                                             values='Total',
                                                                                             aggfunc='sum').rename(columns={'Total': 'total_otros'})

# Unimos las tablas dinámicas en un solo DataFrame final
df_motivo_viaje = pd.concat([pivot_negocios, pivot_ocio, pivot_otros], axis=1).reset_index()

# Redondear las columnas a dos decimales
df_motivo_viaje['total_negocios'] = df_motivo_viaje['total_negocios'].round(2)
df_motivo_viaje['total_ocio'] = df_motivo_viaje['total_ocio'].round(2)
df_motivo_viaje['total_otros'] = df_motivo_viaje['total_otros'].round(2)

df_motivo_viaje
```

```python
#@title Crear la tabla dinámica para via_acceso:

# Convertir la columna "Total" a numérico, si aún no se ha hecho
via_acceso['Total'] = pd.to_numeric(via_acceso['Total'].str.replace('.', '').str.replace(',', '.'), errors='coerce')

# Crear tablas dinámicas para cada tipo de vía de acceso
pivot_aeropuerto = via_acceso[via_acceso['Vía de acceso'] == 'Aeropuerto'].pivot_table(index=['Año', 'Region'],
                                                                                       values='Total',
                                                                                       aggfunc='sum').rename(columns={'Total': 'total_aeropuerto'})

pivot_carretera = via_acceso[via_acceso['Vía de acceso'] == 'Carretera'].pivot_table(index=['Año', 'Region'],
                                                                                     values='Total',
                                                                                     aggfunc='sum').rename(columns={'Total': 'total_carretera'})

pivot_puerto = via_acceso[via_acceso['Vía de acceso'] == 'Puerto'].pivot_table(index=['Año', 'Region'],
                                                                               values='Total',
                                                                               aggfunc='sum').rename(columns={'Total': 'total_puerto'})

pivot_tren = via_acceso[via_acceso['Vía de acceso'] == 'Tren'].pivot_table(index=['Año', 'Region'],
                                                                           values='Total',
                                                                           aggfunc='sum').rename(columns={'Total': 'total_tren'})

# No necesitamos crear una tabla dinámica para "Total de vías de acceso" ya que es un agregado de los otros

# Unimos las tablas dinámicas en un solo DataFrame final
df_via_acceso = pd.concat([pivot_aeropuerto, pivot_carretera, pivot_puerto, pivot_tren], axis=1).reset_index()

# Redondear las columnas a dos decimales, si es necesario
df_via_acceso = df_via_acceso.round(2)

df_via_acceso
```

```python

#@title Crear la tabla dinámica para duracion_viaje

# Convertir la columna "Total" a numérico, manejando correctamente los separadores
duracion_viaje['Total'] = pd.to_numeric(duracion_viaje['Total'].str.replace('.', '').str.replace(',', '.'), errors='coerce')

# Crear tablas dinámicas para cada categoría de duración del viaje
pivot_1_noche = duracion_viaje[duracion_viaje['Duración del viaje'] == '1 noche'].pivot_table(index=['Año', 'Region'],
                                                                                             values='Total',
                                                                                             aggfunc='sum').rename(columns={'Total': 'total_1_noche'})

pivot_2_3_noches = duracion_viaje[duracion_viaje['Duración del viaje'] == 'De 2 a 3 noches'].pivot_table(index=['Año', 'Region'],
                                                                                                       values='Total',
                                                                                                       aggfunc='sum').rename(columns={'Total': 'total_2_3_noches'})

pivot_4_7_noches = duracion_viaje[duracion_viaje['Duración del viaje'] == 'De 4 a 7 noches'].pivot_table(index=['Año', 'Region'],
                                                                                                       values='Total',
                                                                                                       aggfunc='sum').rename(columns={'Total': 'total_4_7_noches'})

pivot_8_15_noches = duracion_viaje[duracion_viaje['Duración del viaje'] == 'De 8 a 15 noches'].pivot_table(index=['Año', 'Region'],
                                                                                                         values='Total',
                                                                                                         aggfunc='sum').rename(columns={'Total': 'total_8_15_noches'})

pivot_mas_15_noches = duracion_viaje[duracion_viaje['Duración del viaje'] == 'Más de 15 noches'].pivot_table(index=['Año', 'Region'],
                                                                                                           values='Total',
                                                                                                           aggfunc='sum').rename(columns={'Total': 'total_mas_15_noches'})

# Unimos las tablas dinámicas en un solo DataFrame final
df_duracion_viaje = pd.concat([pivot_1_noche, pivot_2_3_noches, pivot_4_7_noches, pivot_8_15_noches, pivot_mas_15_noches], axis=1).reset_index()

# Redondear las columnas a dos decimales
df_duracion_viaje['total_1_noche'] = df_duracion_viaje['total_1_noche'].round(2)
df_duracion_viaje['total_2_3_noches'] = df_duracion_viaje['total_2_3_noches'].round(2)
df_duracion_viaje['total_4_7_noches'] = df_duracion_viaje['total_4_7_noches'].round(2)
df_duracion_viaje['total_8_15_noches'] = df_duracion_viaje['total_8_15_noches'].round(2)
df_duracion_viaje['total_mas_15_noches'] = df_duracion_viaje['total_mas_15_noches'].round(2)

df_duracion_viaje
```

```python
#@title Crear la tabla dinámica para tipo_alojamiento


# Convertir la columna "Total" a numérico, si aún no se ha hecho
tipo_alojamiento['Total'] = pd.to_numeric(tipo_alojamiento['Total'].str.replace('.', '').str.replace(',', '.'), errors='coerce')

# Crear tablas dinámicas para cada tipo de alojamiento
pivot_de_mercado = tipo_alojamiento[tipo_alojamiento['Tipo de alojamiento'] == 'De mercado'].pivot_table(index=['Año', 'Region'],
                                                                                                        values='Total',
                                                                                                        aggfunc='sum').rename(columns={'Total': 'total_de_mercado'})

pivot_hotelero = tipo_alojamiento[tipo_alojamiento['Tipo de alojamiento'] == 'Hotelero'].pivot_table(index=['Año', 'Region'],
                                                                                                     values='Total',
                                                                                                     aggfunc='sum').rename(columns={'Total': 'total_hotelero'})

pivot_no_de_mercado = tipo_alojamiento[tipo_alojamiento['Tipo de alojamiento'] == 'No de mercado'].pivot_table(index=['Año', 'Region'],
                                                                                                               values='Total',
                                                                                                               aggfunc='sum').rename(columns={'Total': 'total_no_de_mercado'})

pivot_resto_de_mercado = tipo_alojamiento[tipo_alojamiento['Tipo de alojamiento'] == 'Resto de mercado'].pivot_table(index=['Año', 'Region'],
                                                                                                                      values='Total',
                                                                                                                      aggfunc='sum').rename(columns={'Total': 'total_resto_de_mercado'})

pivot_resto_de_no_mercado = tipo_alojamiento[tipo_alojamiento['Tipo de alojamiento'] == 'Resto de no mercado'].pivot_table(index=['Año', 'Region'],
                                                                                                                            values='Total',
                                                                                                                            aggfunc='sum').rename(columns={'Total': 'total_resto_de_no_mercado'})

pivot_vivienda_familiares_amigos = tipo_alojamiento[tipo_alojamiento['Tipo de alojamiento'] == 'Vivienda de familiares o amigos'].pivot_table(index=['Año', 'Region'],
                                                                                                                                             values='Total',
                                                                                                                                             aggfunc='sum').rename(columns={'Total': 'total_vivienda_familiares_amigos'})

pivot_vivienda_alquiler = tipo_alojamiento[tipo_alojamiento['Tipo de alojamiento'] == 'Vivienda en alquiler'].pivot_table(index=['Año', 'Region'],
                                                                                                                           values='Total',
                                                                                                                           aggfunc='sum').rename(columns={'Total': 'total_vivienda_alquiler'})

pivot_vivienda_propiedad = tipo_alojamiento[tipo_alojamiento['Tipo de alojamiento'] == 'Vivienda en propiedad'].pivot_table(index=['Año', 'Region'],
                                                                                                                             values='Total',
                                                                                                                             aggfunc='sum').rename(columns={'Total': 'total_vivienda_propiedad'})

# Unimos las tablas dinámicas en un solo DataFrame final
df_tipo_alojamiento= pd.concat([pivot_de_mercado, pivot_hotelero, pivot_no_de_mercado, pivot_resto_de_mercado, pivot_resto_de_no_mercado,
                                       pivot_vivienda_familiares_amigos, pivot_vivienda_alquiler, pivot_vivienda_propiedad], axis=1).reset_index()

# Redondear las columnas a dos decimales, si es necesario
df_tipo_alojamiento_ = df_tipo_alojamiento.round(2)

df_tipo_alojamiento

```
