```python
#@title Crear la tabla dinámica para gastos turistas
# # Convertir la columna "Total" a numérico, asumiendo que los datos ya están en el DataFrame `gastos_turistas`
gastos_turistas['Total'] = pd.to_numeric(gastos_turistas['Total'].str.replace('.', '').str.replace(',', '.'), errors='coerce')

# Primero, filtramos para cada tipo y creamos las columnas correspondientes.

df_gasto_total = gastos_turistas[gastos_turistas['Gastos y duración media de los viajes'] == 'Gasto total']

df_gasto_medio_diario_persona = gastos_turistas[gastos_turistas['Gastos y duración media de los viajes'] == 'Gasto medio diario por persona']

df_gasto_medio_persona = gastos_turistas[gastos_turistas['Gastos y duración media de los viajes'] == 'Gasto medio diario por persona']

df_duracion_media_viajes = gastos_turistas[gastos_turistas['Gastos y duración media de los viajes'] == 'Duración media de los viajes']

# Ahora, creamos la tabla dinámica para cada tipo.

pivot_gasto_total = df_gasto_total.pivot_table(index=['Año','País de residencia',	'Total Nacional y CCAA'],
                                           values='Total',
                                           aggfunc='sum').rename(columns={'Total': 'gasto_total'})

pivot_gasto_medio_diario_persona = df_gasto_medio_diario_persona.pivot_table(index=['Año','País de residencia',	'Total Nacional y CCAA'],
                                           values='Total',
                                           aggfunc='sum').rename(columns={'Total': 'gasto_medio_diario_persona'})

pivot_gasto_medio_persona = df_gasto_medio_persona.pivot_table(index=['Año','País de residencia',	'Total Nacional y CCAA'],
                                           values='Total',
                                           aggfunc='sum').rename(columns={'Total': 'gasto_medio_persona'})

pivot_duracion_media_viajes = df_duracion_media_viajes.pivot_table(index=['Año','País de residencia',	'Total Nacional y CCAA'],
                                           values='Total',
                                           aggfunc='sum').rename(columns={'Total': 'duracion_media_viajes'})




# Unimos las cuatro tablas dinámicas en una sola DataFrame final#usamos pd.concat al combinar varias tablas
df_gastos_turistas   = pd.concat([
    pivot_gasto_total, pivot_gasto_medio_diario_persona,pivot_gasto_medio_persona,pivot_duracion_media_viajes
], axis=1).reset_index()


# Redondear la columna 'df_gastos_turistas' a dos decimales
df_gastos_turistas['gasto_total'] = df_gastos_turistas['gasto_total'].round(2)

df_gastos_turistas['gasto_medio_diario_persona'] = df_gastos_turistas['gasto_medio_diario_persona'].round(2)

df_gastos_turistas['gasto_medio_persona'] = df_gastos_turistas['gasto_medio_persona'].round(2)
df_gastos_turistas['duracion_media_viajes'] = df_gastos_turistas['duracion_media_viajes'].round(2)


df_gastos_turistas
```

```python
#@title Crear la tabla dinámica para viviendas_turistas
# Convertir la columna "Total" a numérico, asumiendo que los datos ya están en el DataFrame `viviendas_turistas
viviendas_turistas['Total'] = pd.to_numeric(viviendas_turistas['Total'].str.replace('.', '').str.replace(',', '.'), errors='coerce')

# Primero, filtramos para cada tipo y creamos las columnas correspondientes.

df_plazas = viviendas_turistas[viviendas_turistas['Viviendas y plazas'] == 'Plazas']

df_plazasxviviendas = viviendas_turistas[viviendas_turistas['Viviendas y plazas'] == 'Plazas por vivienda turística']

df_viviendas_turistica = viviendas_turistas[viviendas_turistas['Viviendas y plazas'] == 'Viviendas turísticas']


# Ahora, creamos la tabla dinámica para cada tipo.

pivot_plazas = df_plazas.pivot_table(index=['Año','Total Nacional y CCAA','Provincias'],
                                           values='Total',
                                           aggfunc='sum').rename(columns={'Total': 'plazas'})

pivot_plazxviviendas = df_plazasxviviendas.pivot_table(index=['Año','Total Nacional y CCAA','Provincias'],
                                           values='Total',
                                           aggfunc='sum').rename(columns={'Total': 'plazasxviviendas'})

pivot_viviendas_turistica = df_viviendas_turistica.pivot_table(index=['Año','Total Nacional y CCAA','Provincias'],
                                           values='Total',
                                           aggfunc='sum').rename(columns={'Total': 'viviendas_turistica'})



# Unimos las cuatro tablas dinámicas en una sola DataFrame final#usamos pd.concat al combinar varias tablas
df_viviendas_turistas  = pd.concat([
    pivot_plazas, pivot_plazxviviendas,pivot_viviendas_turistica,
], axis=1).reset_index()


# Redondear la columna 'df_gastos_turistas' a dos decimales
df_viviendas_turistas['plazas'] = df_viviendas_turistas['plazas'].round(2)

df_viviendas_turistas['plazasxviviendas'] = df_viviendas_turistas['plazasxviviendas'].round(2)

df_viviendas_turistas['viviendas_turistica'] = df_viviendas_turistas['viviendas_turistica'].round(2)

```
